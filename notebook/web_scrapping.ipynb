```python
#Load page


import cloudscraper #Ocupo sloudscraper ya que se me hizo muy dificil acceder a la página por restricciones antibots de cloudfare
from bs4 import BeautifulSoup

url = "https://www.32dientes.com/las-mejores-bicis-de-enduro/"

scraper = cloudscraper.create_scraper()

response = scraper.get(url)
soup = BeautifulSoup(response.text, "html.parser")
soup
```

```python
table_bikes = soup.find_all("table", class_="has-fixed-layout")
print(f'Number of tables {len(table_bikes)}')
```

```python
rows = table_bikes[0].find_all("tr") #Buscamos el tag tr en la primer fila de la tabla
headers_table = rows[0] #Asignamos las etiquetas a los headers
headers_cells = headers_table.find_all("th") #buscamos las etiquetas th en los headers que son los que contienen el texto
headers_data = [] #Inicializamos la lista para asignar los valores de etiqueta a datos en python
headers_data = [cell.get_text(strip=True) for cell in headers_cells] #asignamos cada elemento de las etiquetas cortando por espacios con strip
headers_data
```

```python
content_cells = rows[1:]
content_data = []
for i, data_cell in enumerate(content_cells):
    cells = data_cell.find_all('td')
    row_data = [cell.get_text(strip=True).replace("mm.","").replace("º","").replace("°","") for cell in cells]
    content_data.append(row_data)
content_data
```

```python
df = pd.DataFrame(content_data, columns=headers_data)
df
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df.columns = df.columns.str.replace(".","_")
```

```python
columns_int = ["rec_del_","rec_tras_","reach"]
df[columns_int] = df[columns_int].astype("int")
df["ang_dir_"] = df["ang_dir_"].str.replace(",",".")
df["ang_dir_"] = df["ang_dir_"].astype("float")

df
```
